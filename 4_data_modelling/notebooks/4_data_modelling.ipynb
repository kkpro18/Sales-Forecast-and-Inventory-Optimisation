```python
## ETS Decomposition

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt


# get error trend seasonality (ETS) for time series

result = seasonal_decompose(data['Units Sold'], model='multiplicative')
result.plot()


```
